```python
# Všechny funkce bez ohledu na volební období
f_all = Funkce(volebni_obdobi=-1)
```

```python
# TODO: Proč se tato nekonzistence nepropaguje do Funkce s definovaným volebním obdobím?
# Prozkoumejme nekonzistenci, která vzniká při spojení tabulek 'funkce' a 'typ_funkce' v třídě Funkce.
# Problém souvisí s tím, že obě tabulky mají za rodiče 'typ_organu'.
# Tabulka 'funkce' se merguje s 'organy' přes sloupec 'id_organu', tabulka 'organy' se merguje s 'typ_organu' přes sloupec 'id_typ_organu'.
# Tabulka funkce se merguje s 'typ_funkce' přes sloupec 'id_typ_funkce', tabulka 'typ_funkce' se merguje s 'typ_organu' přes sloupec 'id_typ_organu'.
# Příslušné 'id_typ_organu' a 'id_typ_organu' by se měly pro každý řádek shodovat. Pokud se liší, jedná se o vnitřní nekonzistenci dat.

# Ověřme, že pro "konzistentní řádky" je příslušný 'id_typ_organu' == 'id_typ_org'.
x = pd.merge(left=f_all.funkce, right=f_all.typ_funkce, on='id_typ_funkce', suffixes=('', '__typ_funkce'), how="left")
konzistentni_radky = x[x.nazev_typ_organu_cz.eq(x.nazev_typ_organu_cz__typ_funkce)]
if len(konzistentni_radky) > 0:
    s = konzistentni_radky.sample()
    test = s.iloc[0]
    id_organu = test.id_organu
    id_typ_funkce = test.id_typ_funkce
    id_typ_organu_dle_id_organu = f_all.organy[f_all.organy.id_organu == id_organu].iloc[0].id_typ_organu 
    id_typ_organu_dle_id_typ_funkce = f_all.typ_funkce[f_all.typ_funkce.id_typ_funkce == id_typ_funkce].iloc[0].id_typ_organu
    print(f"Pro 'konzistentní řádky' (např. řádek {s.index[0]}) by měla nastat rovnost: {id_typ_organu_dle_id_organu} = {id_typ_organu_dle_id_typ_funkce}.")

# Pro "nekonzistentní řádky" je naopak příslušný 'id_typ_organu' != 'id_typ_organu'.
nekonzistentni_radky = x[x.nazev_typ_organu_cz.ne(x.nazev_typ_organu_cz__typ_funkce)]
if len(nekonzistentni_radky) > 0:
    s = nekonzistentni_radky.sample()
    test = s.iloc[0]
    id_organu = test.id_organu
    id_typ_funkce = test.id_typ_funkce
    id_typ_organu_dle_id_organu = f_all.organy[f_all.organy.id_organu == id_organu].iloc[0].id_typ_organu 
    id_typ_organu_dle_id_typ_funkce = f_all.typ_funkce[f_all.typ_funkce.id_typ_funkce == id_typ_funkce].iloc[0].id_typ_organu
    print(f"Pro 'nekonzistentní řádky' (např. řádek {s.index[0]}) by měla nastat nerovnost: {id_typ_organu_dle_id_organu} != {id_typ_organu_dle_id_typ_funkce}.")

    r = len(nekonzistentni_radky)/len(x)
    print(f"Protože je nekonzistence velmi malá {r:.5f} , rozhodli jsme se ji ignorovat.")
```
