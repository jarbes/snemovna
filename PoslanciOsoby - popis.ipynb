```python
# Všechny funkce bez ohledu na volební období
f_all = Funkce(volebni_obdobi=-1)
```

```python
# TODO: Proč se tato nekonzistence nepropaguje do Funkce s definovaným volebním obdobím?
# Prozkoumejme nekonzistenci, která vzniká při spojení tabulek 'funkce' a 'typ_funkce' v třídě Funkce.
# Problém souvisí s tím, že obě tabulky mají za rodiče 'typ_organu'.
# Tabulka 'funkce' se merguje s 'organy' přes sloupec 'id_organu', tabulka 'organy' se merguje s 'typ_organu' přes sloupec 'id_typ_organu'.
# Tabulka funkce se merguje s 'typ_funkce' přes sloupec 'id_typ_funkce', tabulka 'typ_funkce' se merguje s 'typ_organu' přes sloupec 'id_typ_organu'.
# Příslušné 'id_typ_organu' a 'id_typ_organu' by se měly pro každý řádek shodovat. Pokud se liší, jedná se o vnitřní nekonzistenci dat.

# Ověřme, že pro "konzistentní řádky" je příslušný 'id_typ_organu' == 'id_typ_org'.
x = pd.merge(left=f_all.tbl['funkce'], right=f_all.tbl['typ_funkce'], on='id_typ_funkce', suffixes=('', '__typ_funkce'), how="left")
konzistentni_radky = x[x.nazev_typ_organu_cz.eq(x.nazev_typ_organu_cz__typ_funkce)]
if len(konzistentni_radky) > 0:
    s = konzistentni_radky.sample()
    test = s.iloc[0]
    id_organu = test.id_organu
    id_typ_funkce = test.id_typ_funkce
    id_typ_organu_dle_organu = f_all.tbl['organy'][f_all.tbl['organy'].id_organu == id_organu].iloc[0].id_typ_organu 
    id_typ_organu_dle_typu_funkce = f_all.tbl['typ_funkce'][f_all.tbl['typ_funkce'].id_typ_funkce == id_typ_funkce].iloc[0].id_typ_organu
    print(f"Pro 'konzistentní řádky' (např. řádek {s.index[0]}) by měla nastat rovnost: {id_typ_organu_dle_organu} = {id_typ_organu_dle_typu_funkce}.")

# Pro "nekonzistentní řádky" je naopak příslušný 'id_typ_organu' != 'id_typ_organu'.
nekonzistentni_radky = x[x.nazev_typ_organu_cz.ne(x.nazev_typ_organu_cz__typ_funkce)]
if len(nekonzistentni_radky) > 0:
    s = nekonzistentni_radky.sample()
    test = s.iloc[0]
    #print(f"nekonzistentni radek: {test}")
    id_organu = test.id_organu
    id_typ_funkce = test.id_typ_funkce
    id_typ_organu_dle_organu = f_all.tbl['organy'][f_all.tbl['organy'].id_organu == id_organu].iloc[0].id_typ_organu 
    id_typ_organu_dle_typu_funkce = f_all.tbl['typ_funkce'][f_all.tbl['typ_funkce'].id_typ_funkce == id_typ_funkce].iloc[0].id_typ_organu
    print(f"Pro 'nekonzistentní řádky' (např. řádek {s.index[0]}) by měla nastat nerovnost: {id_typ_organu_dle_organu} != {id_typ_organu_dle_typu_funkce}.")

    r = 100*len(nekonzistentni_radky)/len(x)
    print(f"Protože je nekonzistence velmi malá ({r:.3f}%), rozhodli jsme se ji ignorovat.")
```

## Další analýza

```python
vsechny_organy = Organy(volebni_obdobi=-1)
snemovny = vsechny_organy[vsechny_organy.nazev_typ_organu_cz == 'Parlament'].od_organ.dt.year.sort_values()

poslanci = {}
for i in snemovny:
    poslanci[i] = Poslanci(volebni_obdobi=i)
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=1)

for i in snemovny:
    p = poslanci[i].groupby('pohlavi').size()
    total = len(p)
    text = [str(i), str(i)]
    fig.add_trace(go.Bar(
            x=p.index,
            y=p.values,
            text=text,
            textposition="inside",
            name=f"sněmovna {i}",
            hovertemplate=f"Sněmovna: {i}" +'<br>Pohlaví: %{x}<br>Počet: %{y}<extra></extra>'),
        row=1, col=1)

layout = go.Layout(
    title="Počty poslanců ve sněmovně dle pohlaví",    
    plot_bgcolor="#FFFFFF",
    #hovermode="x",
    #hoverdistance=100, # Distance to show hover label of data point
    #spikedistance=1000, # Distance to show spike
    xaxis=dict(
        title="pohlaví",
        linecolor="#BCCCDC",
    ),
    yaxis=dict(
        title="počet",
        linecolor="#BCCCDC",
        showspikes=True,
        spikethickness=1,
        spikedash="dot",
        spikecolor="#999999",
        spikemode="across",
    )
)

fig.update_layout(layout)
fig.show()
```

```python
#id_kandidatka -> strana
from datetime import datetime

fig = make_subplots(rows=3, cols=len(snemovny))
for idx, i in enumerate(snemovny):
    p = poslanci[i]
    p['vek'] = (p.snemovna.od_organ - p.narozeni) / pd.Timedelta(days=365)
    fig.add_trace(go.Box(y=p.vek, name=f"{i}"), row=1, col=idx+1)
    fig.add_trace(go.Box(y=p[p.pohlavi=='muž'].vek, name=f"{i} - muži"), row=2, col=idx+1)
    fig.add_trace(go.Box(y=p[p.pohlavi=='žena'].vek, name=f"{i} - ženy"), row=3, col=idx+1)

fig.show()
```

```python
#id_kandidatka -> strana
from datetime import datetime

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[i for i in snemovny],
    y=[poslanci[i].vek.mean() for i in snemovny],
    mode="markers+lines",
    name=f"všichni")
)

fig.add_trace(go.Scatter(
    x=[i for i in snemovny],
    y=[poslanci[i][poslanci[i].pohlavi=='muž'].vek.mean() for i in snemovny],
    mode="markers+lines",
    name=f"muži")
)

fig.add_trace(go.Scatter(
    x=[i for i in snemovny],
    y=[poslanci[i][poslanci[i].pohlavi=='žena'].vek.mean() for i in snemovny],
    mode="markers+lines",
    name=f"ženy")
)

layout = go.Layout(
    title="Průměrný věk poslankyň a poslanců při nástupu do funkce pro různé sněmovny",    
    plot_bgcolor="#FFFFFF",
    #hovermode="x",
    #hoverdistance=100, # Distance to show hover label of data point
    #spikedistance=1000, # Distance to show spike
    xaxis=dict(
        title="sněmovna",
        linecolor="#BCCCDC",
        type="category"
    ),
    yaxis=dict(
        title="průměrný věk",
        linecolor="#BCCCDC",
        showspikes=True,
        spikethickness=1,
        spikedash="dot",
        spikecolor="#999999",
        spikemode="across",
    )
)

fig.update_layout(layout)
fig.show()
```
