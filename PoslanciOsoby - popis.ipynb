```python
# Všechny funkce bez ohledu na volební období
f_all = Funkce(volebni_obdobi=-1)
```

```python
# TODO: Proč se tato nekonzistence nepropaguje do Funkce s definovaným volebním obdobím?
# Prozkoumejme nekonzistenci, která vzniká při spojení tabulek 'funkce' a 'typ_funkce' v třídě Funkce.
# Problém souvisí s tím, že obě tabulky mají za rodiče 'typ_organu'.
# Tabulka 'funkce' se merguje s 'organy' přes sloupec 'id_organu', tabulka 'organy' se merguje s 'typ_organu' přes sloupec 'id_typ_organu'.
# Tabulka funkce se merguje s 'typ_funkce' přes sloupec 'id_typ_funkce', tabulka 'typ_funkce' se merguje s 'typ_organu' přes sloupec 'id_typ_organu'.
# Příslušné 'id_typ_organu' a 'id_typ_organu' by se měly pro každý řádek shodovat. Pokud se liší, jedná se o vnitřní nekonzistenci dat.

# Ověřme, že pro "konzistentní řádky" je příslušný 'id_typ_organu' == 'id_typ_org'.
x = pd.merge(left=f_all.funkce, right=f_all.typ_funkce, on='id_typ_funkce', suffixes=('', '__typ_funkce'), how="left")
konzistentni_radky = x[x.nazev_typ_organu_cz.eq(x.nazev_typ_organu_cz__typ_funkce)]
if len(konzistentni_radky) > 0:
    s = konzistentni_radky.sample()
    test = s.iloc[0]
    id_organu = test.id_organu
    id_typ_funkce = test.id_typ_funkce
    id_typ_organu_dle_id_organu = f_all.organy[f_all.organy.id_organu == id_organu].iloc[0].id_typ_organu 
    id_typ_organu_dle_id_typ_funkce = f_all.typ_funkce[f_all.typ_funkce.id_typ_funkce == id_typ_funkce].iloc[0].id_typ_organu
    print(f"Pro 'konzistentní řádky' (např. řádek {s.index[0]}) by měla nastat rovnost: {id_typ_organu_dle_id_organu} = {id_typ_organu_dle_id_typ_funkce}.")

# Pro "nekonzistentní řádky" je naopak příslušný 'id_typ_organu' != 'id_typ_organu'.
nekonzistentni_radky = x[x.nazev_typ_organu_cz.ne(x.nazev_typ_organu_cz__typ_funkce)]
if len(nekonzistentni_radky) > 0:
    s = nekonzistentni_radky.sample()
    test = s.iloc[0]
    id_organu = test.id_organu
    id_typ_funkce = test.id_typ_funkce
    id_typ_organu_dle_id_organu = f_all.organy[f_all.organy.id_organu == id_organu].iloc[0].id_typ_organu 
    id_typ_organu_dle_id_typ_funkce = f_all.typ_funkce[f_all.typ_funkce.id_typ_funkce == id_typ_funkce].iloc[0].id_typ_organu
    print(f"Pro 'nekonzistentní řádky' (např. řádek {s.index[0]}) by měla nastat nerovnost: {id_typ_organu_dle_id_organu} != {id_typ_organu_dle_id_typ_funkce}.")

    r = len(nekonzistentni_radky)/len(x)
    print(f"Protože je nekonzistence velmi malá {r:.5f} , rozhodli jsme se ji ignorovat.")
```

```python
# Jak se překrývá konec předchozí sněmovny a začátek nové sněmovny?

snemovny = org.df[org.df.nazev_typ_organu_cz == 'Parlament'].sort_values(by="od_organ").copy()
snemovny['do_organ_predchozi'] = snemovny['do_organ'].shift(1)
snemovny[snemovny.do_organ_predchozi > snemovny.od_organ]
```

```python
# Najdi sněmovní orgány
def find_children_ids(ids, id_field, df, parent_field, parent_id, level):
    children = df[df[parent_field] == parent_id]
    if len(children) > 0:
        for idx, child in children.iterrows():
            #print(' '*level, ' ', child.nazev_organu_cz)
            ids.append(child[id_field])
            ids = find_children_ids(ids, id_field, df, parent_field, child.id_organu, level+1)
    return ids

print(f"Children of '{org.df[org.df.id_organu == org.id_organu].iloc[0].zkratka}':")
ids = [org.id_organu]
ids = find_children_ids(ids, 'id_organu', org.df, 'organ_id_organu', org.id_organu, 0)
#org.df['patri_do_snemovny'] 
org.df['je_snemovni_organ'] = org.df.id_organu.isin(ids) == True
org.df[org.df['je_snemovni_organ'] == True].set_index('id_organu').reindex(ids).reset_index()
```

```python
# TODO: Kdy použít od_f místo od_o, resp. do_f místo do_o?

interval_start = org.df.od_organ.mask(org.df.od_organ.isna(), org.volebni_obdobi_od).mask(~org.df.od_organ.isna(), np.maximum(org.df.od_organ, org.volebni_obdobi_od))

# Pozorování: volebni_obdobi_od není nikdy NaT => interval_start není nikdy NaT  
if pd.isna(org.volebni_obdobi_do): # příznak posledního volebního období
    podminka = (
        (interval_start.dt.date <= org.df.do_organ.dt.date) # Nutná podmínka pro True: (interval_start != NaT, splněno vždy) a (do_organ != NaT)  
        |  (org.df.do_organ.isna()) # Nutná podmínka pro True: (interval_start != NaT, splněno vždy) a (do_organ == NaT)
    )
else: # Pozorování: předchozí volební období => interval_end není nikdy NaT
    interval_end = org.df.do_organ.mask(org.df.do_organ.isna(), org.volebni_obdobi_do).mask(~org.df.do_organ.isna(), np.minumum(org.df.do_organ, org.volebni_obdobi_do))
    podminka = (interval_start.dt.date <= interval_end.dt.date)

organy_snemovny = org.df[podminka | org.df.je_snemovni_organ == True]
#organy_snemovny[organy_snemovny.nazev_organu_cz == 'Poslanecká sněmovna']
organy_snemovny[organy_snemovny.do_organ.isna() | (organy_snemovny.do_organ > "2017-10-26")].sort_values(by="do_organ") # Orgány snemovny

organy_snemovny[organy_snemovny.do_organ <= "2017-10-26"].sort_values(by="do_organ") # S koncem předchozí sněmovny také končí předchozí orgány
```
